#1. 다음을 출력하는 프로그램을 작성하시오.

> **100** little **monkeys** jumping on the bed.

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, "No more monkeys jumping on the bed!"

> ...

> ... 

> **1** little **monkey** jumping on the bed. 

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, 

> "No more monkeys jumping on the bed!" 

(참고) Mama 혹은 Daddy를 랜돔하게 선택하게 하자.

```python
#100, Monkeys, his(her), Mama(Daddy) 가 바뀌는 부분 his(her), Mama(Daddy)는 random choice가 되어야함

import random
list1=["her", "his"]
list2=["Mama", "Daddy"]
multiline_str = """%s little %s jumping on the bed.\n
One fell off and broke %s head.\n 
%s called the doctor and the doctor said,\n
\"No more monkeys jumping on the bed!\"\n\n"""

for x in range(100):
    if x == 99:
        numMonkey = 100 - x
        mainMonkey = "monkey"
        ownerMonkey = random.choice(list1) 
        parentMonkey = random.choice(list2)
    else:
        numMonkey = 100 - x
        mainMonkey = "monkeys"
        ownerMonkey = random.choice(list1) 
        parentMonkey = random.choice(list2)
        
    print(multiline_str % (numMonkey,mainMonkey,ownerMonkey,parentMonkey))
```

#2. 시작하는 연도와 끝나는 연도를 물어보는 프로그램을 작성 하고, 그 사이에 있는 모든 윤년을 출력하라. 

(시작하는 연도와 끝나는 연도가 윤년이면 그 해도 출력) 

* 윤년은 4로 나누어짐 (예: 1984, 2012)
* 100으로 나누어지면 윤년이 아님 (예: 1800, 1900) 
* 윤년은 400으로 나누어짐 (예: 1600, 2000) 

> Pick a starting year (like 1973 or something): 1973

> Now pick an ending year: 1977 

> Check it out... these years are leap years: 1976 

(참고) 윤년 리스트: http://mwultong.blogspot.com/2005/12/18002300-leap-year-list.html

```python
start_year = input("Pick a starting year (like 1973 or something): ")
end_year = input("Now pick an ending year: ")
yun_year=[]
for x in range(int(start_year), int(end_year)+1):
    if x%4 == 0 and not(x%100 == 0 and x%400 != 0):
        yun_year.append(x)
        
print("Check it out... these years are leap years: %s" %(str(yun_year)[1:-1]))
```

#3. 구구단을 출력하는 프로그램을 작성하고 결과를 txt 파일로 저장하시오.

```python
f = open("gugudan.txt", "w", encoding="utf-8")

for x in range (1,10):
    for y in range (0,10):
        if y == 0:
            f.write(str(x)+"단\n")
        else:
            f.write("%s x %s = %s\n" % (x, y, x*y))
f.close()
```
